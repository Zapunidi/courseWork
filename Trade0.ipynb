```python
class Environment:
    DF_PATH = 'AAPL.csv'
    INITIAL_MONEY = 10000
    INITIAL_STOCKS = []
    
    
    def __init__(self):
        self.df = pd.read_csv(self.DF_PATH)
        if 'Date' in self.df:
            self.df.pop('Date')
        self.data = self.df.values
        self.current_money = self.INITIAL_MONEY
        self.current_stocks = self.INITIAL_STOCKS.copy()
        self.cursor = 0
        
    def reset(self):
        self.cursor = 0
        
    def sell(self):
        if len(self.current_stocks) > 0:
            bought = self.current_stocks.pop(0)
            sold = self.data[self.cursor][2]
            profit = sold - bought
        else:
            profit = 0
        return profit
    
    def hold(self):
        return 0
    
    def buy(self):
        if self.current_money >= self.data[self.cursor][1]:
            self.current_money -= self.data[self.cursor][1]
            self.current_stocks.append(self.data[self.cursor][1])
            self.current_stocks.sort()
        return 0
            
    def reward(self, profit):
        return profit + self.current_money / self.INITIAL_MONEY
    
    def get_state(self):
        state = self.data[self.cursor]
        return state
        
    def step(self, action):
        done = False
        self.cursor += 1
        
        if action == -1:
            profit = self.sell()
        elif action == 0:
            profit = self.hold()
        elif action == 1:
            profit = self.buy()
            
        if self.cursor == len(self.data) - 1:
            done = True
            
        return self.get_state(), self.reward(profit), done
```

```python
import sys
import gym
import pylab
import numpy as np
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

EPISODES = 100


# A2C(Advantage Actor-Critic) agent for the Cartpole
class A2CAgent:
    def __init__(self, state_size, action_size):
        # if you want to see Cartpole learning, then change to True
        self.render = False
        self.load_model = False
        # get size of state and action
        self.state_size = state_size
        self.action_size = action_size
        self.value_size = 1

        # These are hyper parameters for the Policy Gradient
        self.discount_factor = 0.99
        self.actor_lr = 0.0001
        self.critic_lr = 0.0001

        # create model for policy network
        self.actor = self.build_actor()
        self.critic = self.build_critic()

        if self.load_model:
            self.actor.load_weights("./save_model/cartpole_actor.h5")
            self.critic.load_weights("./save_model/cartpole_critic.h5")

    # approximate policy and value using Neural Network
    # actor: state is input and probability of each action is output of model
    def build_actor(self):
        actor = Sequential()
        actor.add(Dense(32, input_dim=self.state_size, activation='relu',
                        kernel_initializer='he_uniform'))
        actor.add(Dense(self.action_size, activation='softmax',
                        kernel_initializer='he_uniform'))
        actor.summary()
        # See note regarding crossentropy in cartpole_reinforce.py
        actor.compile(loss='categorical_crossentropy',
                      optimizer=Adam(lr=self.actor_lr))
        return actor

    # critic: state is input and value of state is output of model
    def build_critic(self):
        critic = Sequential()
        critic.add(Dense(32, input_dim=self.state_size, activation='relu',
                         kernel_initializer='he_uniform'))
        critic.add(Dense(self.value_size, activation='linear',
                         kernel_initializer='he_uniform'))
        critic.summary()
        critic.compile(loss="mse", optimizer=Adam(lr=self.critic_lr))
        return critic

    # using the output of policy network, pick action stochastically
    def get_action(self, state):
        policy = self.actor.predict(state, batch_size=1).flatten()
        return np.random.choice(self.action_size, 1, p=policy)[0]

    # update policy network every episode
    def train_model(self, state, action, reward, next_state, done):
        target = np.zeros((1, self.value_size))
        advantages = np.zeros((1, self.action_size))

        value = self.critic.predict(state)[0]
        next_value = self.critic.predict(next_state)[0]

        if done:
            advantages[0][action] = reward - value
            target[0][0] = reward
        else:
            advantages[0][action] = reward + self.discount_factor * (next_value) - value
            target[0][0] = reward + self.discount_factor * next_value

        self.actor.fit(state, advantages, epochs=1, verbose=0)
        self.critic.fit(state, target, epochs=1, verbose=0)


if __name__ == "__main__":
    en = 0
    comission = 0.01
    start_money = 10000
    money = start_money
    # In case of CartPole-v1, maximum length of episode is 500
    #env = gym.make(envName)
    # get size of state and action from environment
    #state_size = env.observation_space.shape[0]
    #action_size = env.action_space.n
    
    df = pd.read_csv('AAPL.csv')
    if 'Date' in df:
        df.pop('Date')
    state_size = 6
    action_size = 3
    epis = [df[i-100:i] for i in range(100, df.shape[0], 100)]

    # make A2C agent
    agent = A2CAgent(state_size, action_size)

    scores, episodes = [], []

    for e in epis:
        done = False
        score = 0
        n = 0
        state = e.values[n]
        state = np.reshape(state, [1, state_size])
        state = state.astype(float)
        acount = [0, 0, 0]

        while n < 99:

            action = agent.get_action(state)
            #next_state, reward, done, info = env.step(action)
            
            n += 1
            next_state = e.values[n]
            next_state = np.reshape(state, [1, state_size])
            next_state = state.astype(float)
            acount[action] += 1
            action = action - 1
            
            
            
            if action == -1: # sell
                money += state[0][2]
                
            if action == 1: # buy
                money -= state[0][1]
            
            reward = money - start_money

            agent.train_model(state, action, reward, next_state, done=False)

            score += reward
            state = next_state

        
            # every episode, plot the play time
            #score = score if score == 500.0 else score + 100
        print(acount)
        agent.train_model(state, action, reward, next_state, done=True)
        scores.append(score)
        episodes.append(en)
        en+=1
        pylab.plot(episodes, scores, 'b')
        pylab.savefig("cartpole_a2c.png")
        print("episode:", en,  "score:", score)

                # if the mean of scores of last 10 episode is bigger than 490
                # stop training
                #if np.mean(scores[-min(10, len(scores)):]) > -10:
                #    break

        # save the model
        #if e % 25 == 0:
        #agent.actor.save_weights("./save_model/cartpole_actor2.h5")
        #agent.critic.save_weights("./save_model/cartpole_critic2.h5")
```
